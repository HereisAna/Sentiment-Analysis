```python
import pandas as pd
import numpy as np
from textblob import TextBlob
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
#Load the data
df = pd.read_csv('H:\EXCEL\SentimentEvalData.csv', encoding_errors= 'replace')
#Show data
df
```

```python
df.head()
```

```python
#Create a fuction to get the polarity
def getPolarity(text):
    return TextBlob(text).sentiment.polarity

#Creat new column
df['Polarity'] = df['Review'].apply(getPolarity)

#Show the data
df
```

```python
df.describe()
```

```python
#Create a new sentiment column
def getSentiment(score):
    if score < 0:
        return 'negative'
    elif score == 0:
        return 'neutral'
    else:
        return 'postive'
#Create a new column
df['Sentiment'] = df['Polarity'].apply(getSentiment)
#Shoe new data
df
    
```

```python
#Plot and visualize the sentiment count
df['Sentiment'].value_counts().plot(kind = 'bar')
plt.xlabel('Sentiment')
plt.ylabel('Counts')
plt.title('Sentiment Analysis')
plt.show()
```
